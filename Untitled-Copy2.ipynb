```python
import pandas as pd
```

```python
dataset = pd.read_excel("Online Retail.xlsx")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset[dataset['Description'].isnull()][51:100]
```

```python
dataset.dropna(subset=['Description'], inplace=True)
```

```python
dataset['CustomerType'] = dataset['CustomerID'].apply(lambda x: 'Guest' if pd.isnull(x) else 'Registered')
```

```python
dataset.CustomerType.value_counts()
```

```python
dataset['CustomerID'] = dataset['CustomerID'].apply(lambda x: -1 if pd.isna(x) else x)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset = dataset[dataset['UnitPrice'] != 0]
```

```python
dataset.Description.value_counts().sort_index()
```

```python
import numpy as np
```

```python
descriptions = np.unique(dataset.Description.tolist())
```

```python
lower_descriptions = list(map(str.lower, descriptions))
```

```python
dataframe = pd.DataFrame(lower_descriptions)
```

```python
dataframe.to_csv("product_descriptions.csv", index=False, header=False)
```

```python
categories = pd.read_csv("categories.csv", header=None)
categories.head()
```

```python
dataframe.rename(columns={0: "Description"}, inplace=True)
```

```python
dataframe = dataframe.join(categories)
```

```python
dataframe.rename(columns={0: "Category"}, inplace=True)
```

```python
dataframe.head()
```

```python
description_category_dictionary = dict(zip(dataframe['Description'], dataframe['Category']))
```

```python
dataset["Category"] = [description_category_dictionary[i.lower()] for i in dataset['Description']]
```

```python
dataset.tail()
```

```python
dataset.to_csv("Final_Data.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the Seaborn style
sns.set_style("whitegrid")

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8))

# Plot UnitPrice
sns.scatterplot(x=dataset.index, y=dataset.UnitPrice, ax=ax1, color='royalblue')
ax1.set_title('Unit Price Distribution', fontsize=16)
ax1.set_xlabel('Index', fontsize=12)
ax1.set_ylabel('Unit Price', fontsize=12)

# Plot Quantity
sns.scatterplot(x=dataset.index, y=dataset.Quantity, ax=ax2, color='forestgreen')
ax2.set_title('Quantity Distribution', fontsize=16)
ax2.set_xlabel('Index', fontsize=12)
ax2.set_ylabel('Quantity', fontsize=12)

# Add a main title
fig.suptitle('Distribution of Unit Price and Quantity', fontsize=20, y=1.05)

# Adjust layout and display
plt.tight_layout()
plt.show()
```
