```python
data = pd.read_csv('product_descriptions.csv', header=None)
```

```python
descriptions = data[0].to_list()
```

```python
classifier = pipeline('zero-shot-classification', model="facebook/bart-large-mnli")
```

```python
categories = [
    "Home & Living",
    "Gifts & Occasions",
    "Decor & Furniture",
    "Toys & Games",
    "Electronics & Gadgets",
    "Kitchen & Dining",
    "Fashion & Apparel",
    "Outdoor & Garden",
    "Beauty & Personal Care",
    "Party Supplies & Events",
    "Crafts & DIY",
    "Storage & Organization",
    "Stationery & Office Supplies",
    "Seasonal & Holiday",
    "Baby & Kids",
    "Pets & Pet Supplies"
]
```

```python
for i in descriptions[:2]:
    result = classifier(i, categories)
    print(i, ' --> ',result['labels'][0])
```

```python
len(data)
```

```python
batch_size, batch_no = 64, 0
description_results = []
for i in range(0, len(data), batch_size):
  batch = descriptions[i: i+batch_size]
  batch_result = classifier(batch, categories)
  for result in batch_result:
    description_results.append(result['labels'][0])
  print("Batch No -> ", batch_no)
  batch_no += 1
```

```python
for i in range(100):
    print(f"{descriptions[i]}: {description_results[i]}")
```

```python
categories = pd.DataFrame(description_results)
```

```python
categories.to_csv("categories.csv", header=None, index=False)
```
